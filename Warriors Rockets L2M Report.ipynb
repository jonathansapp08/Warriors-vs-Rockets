```python
import numpy as np
import pandas as pd

calls = pd.read_csv('/Users/jonathansapp/code/notebooks/Warriors_Rockets_L2M.csv', delimiter=',')
CNC = 0
INC = 0
CC = 0
IC = 0
rockets = ['Rockets', 'Capela', 'Harden', 'Paul', 'Tucker', 'Gordon', 'Gerald','Rivers']
warriors = ['Warriors', 'Durant', 'Curry', 'Green', 'Thompson', 'Livingston', 'Looney', 'Iguodala']
```

```python
for index, row in calls.iterrows():
    if (row['Decision']) == 'CNC':
        CNC = CNC + 1
    elif (row['Decision']) == 'INC':
        INC = INC + 1
    elif (row['Decision']) == 'CC':
        CC = CC + 1
    elif (row['Decision']) == 'IC':
        IC = IC + 1
```

```python
print(CNC)
print(INC)
print(CC)
print(IC)
```

```python
rockets_calls = 0
warriors_calls = 0
for index, row in calls.iterrows():
        if (row['Disadvantaged Player']) in warriors:
            warriors_calls = warriors_calls + 1
for index, row in calls.iterrows():
        if (row['Disadvantaged Player']) in rockets:
            rockets_calls = rockets_calls + 1            
            
```

```python
print(rockets_calls)
print(warriors_calls)
```

```python
# CNC is not needed because nobody was disadvanted 
# CC LEFT is disadvanteged
# IC LEFT is disadvanteged
# INC RIGHT is disadvanteged


left_calls = calls[~calls.Decision.str.contains('CNC')]
left_calls = left_calls[~left_calls.Decision.str.contains('INC')]


right_calls = calls[~calls.Decision.str.contains('CNC')]
right_calls = right_calls[~right_calls.Decision.str.contains('IC')]
right_calls = right_calls[~right_calls.Decision.str.contains('CC')]
```

```python
right_calls
```

```python
rockets_dis = 0
warriors_dis = 0

for index, row in left_calls.iterrows():
        if (row['Committing Player']) in warriors:
            warriors_dis = warriors_dis + 1
for index, row in left_calls.iterrows():
        if (row['Committing Player']) in rockets:
            rockets_dis = rockets_dis + 1   
            
            
for index, row in right_calls.iterrows():
        if (row['Disadvantaged Player']) in warriors:
            warriors_dis = warriors_dis + 1
for index, row in right_calls.iterrows():
        if (row['Disadvantaged Player']) in rockets:
            rockets_dis = rockets_dis + 1   
```

```python
print(rockets_dis)
print(warriors_dis)
```
